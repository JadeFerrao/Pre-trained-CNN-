```python
import os
import glob

def count_images_in_subdirectories(data_dir, directory_name):
    print(f"{directory_name}:")
    # Get list of subdirectories
    subdirectories = [d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))]

    # Iterate through subdirectories
    for subdirectory in subdirectories:
        # Get list of image files in the subdirectory
        image_files = glob.glob(os.path.join(data_dir, subdirectory, '*.jpg'))
        print(f"Subdirectory {subdirectory}: {len(image_files)} images")
    print()  # Print an empty line for separation

# Example usage
data_dir_train = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/train"
count_images_in_subdirectories(data_dir_train, "Train")

data_dir_test = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/test"
count_images_in_subdirectories(data_dir_test, "Test")

data_dir_val = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/val"
count_images_in_subdirectories(data_dir_val, "Validation")

data_dir_val = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/train_cut"
count_images_in_subdirectories(data_dir_val, "Train-cut")

data_dir_val = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/train_cut2"
count_images_in_subdirectories(data_dir_val, "Train-cut2")
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Data directory
data_dir = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/train_cut2"

# Load InceptionV3 model
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(299, 299, 3))

# Define data generator with rescaling only
datagen = ImageDataGenerator(rescale=1./255)

# Generate data iterator
batch_size = 32
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(299, 299),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Extract features using InceptionV3
X = base_model.predict(train_generator)

# Flatten the features
X_flattened = X.reshape(X.shape[0], -1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_flattened, train_generator.classes, test_size=0.2, random_state=42)

# Fit a RandomForestClassifier which is used for classification n regression tasks
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Predict classes
y_pred = rf_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

# ***MobileNet-V2(SMOTE and Random Forest)***

```python
import numpy as np
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Data directory
data_dir = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/train_cut2"

# Load MobileNetV2 model
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Define data generator with rescaling only
datagen = ImageDataGenerator(rescale=1./255)

# Generate data iterator
batch_size = 32
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),  # MobileNetV2's input size is 224x224
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Extract features using MobileNetV2
X = base_model.predict(train_generator)

# Flatten the features
X_flattened = X.reshape(X.shape[0], -1)

# Apply SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_flattened, train_generator.classes)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Fit a RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Predict classes
y_pred = rf_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

# ***Resnet152V2(Cross validation & Random forest)***

```python
import numpy as np
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import StratifiedKFold
from imblearn.over_sampling import SMOTE
from tensorflow.keras.applications import ResNet152V2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.ensemble import RandomForestClassifier

# Data directory
data_dir = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/train_cut2"

# Load ResNet152V2 model
base_model = ResNet152V2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Define data generator with rescaling only
datagen = ImageDataGenerator(rescale=1./255)

# Generate data iterator
batch_size = 32
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),  # ResNet152V2's input size is 224x224
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Extract features using ResNet152V2
X = base_model.predict(train_generator)

# Flatten the features
X_flattened = X.reshape(X.shape[0], -1)

# Apply SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_flattened, train_generator.classes)

# Perform cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []
for fold, (train_index, test_index) in enumerate(skf.split(X_resampled, y_resampled), 1):
    X_train, X_test = X_resampled[train_index], X_resampled[test_index]
    y_train, y_test = y_resampled[train_index], y_resampled[test_index]

    # Fit a RandomForestClassifier
    rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
    rf_classifier.fit(X_train, y_train)

    # Predict classes
    y_pred = rf_classifier.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

    # Print classification report for the fold
    print(f"Classification Report - Fold {fold}:")
    print(classification_report(y_test, y_pred))
    print("Accuracy - Fold {}: {:.4f}".format(fold, accuracy))

# Print average accuracy
print("Average Accuracy:", np.mean(accuracies))
```

```python
import matplotlib.pyplot as plt

# Plotting accuracy for each fold
plt.figure(figsize=(8, 6))
plt.plot(range(1, 6), accuracies, marker='o', linestyle='-')
plt.title('Accuracy for each fold')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

# ***EfficientNet(Ensemble Methods)***

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, Input
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

# Data directory
data_dir = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/train_cut"

# Define data generator with augmentation
datagen = ImageDataGenerator(
    rotation_range=45,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

# Generate data iterators
batch_size = 32
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Load EfficientNetB0 model
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_tensor=Input(shape=(224, 224, 3)))

# Add custom classification head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(train_generator.num_classes, activation='softmax')(x)

# Create base model
model = Model(inputs=base_model.input, outputs=predictions)

# Compile model
optimizer = Adam(lr=0.0001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Train base model
model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=5,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size
)

# Perform ensemble predictions
ensemble_predictions = []
for i in range(3):  # Ensemble size
    # Train a new model
    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // batch_size,
        epochs=5,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // batch_size
    )
    # Make predictions on validation set
    predictions = model.predict(validation_generator)
    ensemble_predictions.append(predictions)

# Compute ensemble accuracy
ensemble_accuracy = model.evaluate(validation_generator)
print(f"Ensemble Accuracy: {ensemble_accuracy[1]}")
```

```python
import numpy as np
import os
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from imblearn.over_sampling import SMOTE

# Data directory
data_dir = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/train_cut2"

# Set up data generator
image_size = (224, 224)  # EfficientNetB0 input size
batch_size = 32
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

# Model
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Extract features using EfficientNetB0
X = base_model.predict(train_generator)

# Flatten the features
X_flattened = X.reshape(X.shape[0], -1)

# Apply SMOTE to balance dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_flattened, train_generator.classes)

# Split the resampled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Use AdaBoostClassifier as ensemble method technique,ombines multiple weak classifiers to create a strong classifier. It sequentially trains the weak classifiers, adjusting their weights based on misclassified samples, and combines their outputs through a weighted sum to make predictions.
adaboost_classifier = AdaBoostClassifier(n_estimators=100, random_state=42)
adaboost_classifier.fit(X_train, y_train)

# Predict classes
y_pred = adaboost_classifier.predict(X_test)

# Calculate accuracy and other metrics
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:")
print(report)
```

#***InceptionResNetV2 (SMOTE & PCA)***

```python
import numpy as np
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from tensorflow.keras.applications import InceptionResNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Data directory
data_dir = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/train_cut2"

# Load InceptionResNetV2 model
base_model = InceptionResNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Define data generator with rescaling only
datagen = ImageDataGenerator(rescale=1./255)

# Generate data iterator
batch_size = 32
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),  # InceptionResNetV2's input size is 224x224
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Extract features using InceptionResNetV2
X = base_model.predict(train_generator)

# Flatten the features
X_flattened = X.reshape(X.shape[0], -1)

# Apply SMOTE to handle class imbalance
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_flattened, train_generator.classes)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Apply PrincipleComponetAnalysis for dimensionality reduction
pca = PCA(n_components=100, random_state=42)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Fit a RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train_pca, y_train)

# Predict classes
y_pred = rf_classifier.predict(X_test_pca)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

# ***Test with epochs ; all 4 models***

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionResNetV2, Xception, MobileNet, DenseNet121
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report

# Define the directory where your dataset is located
data_dir = "/content/drive/MyDrive/Bigdatasets/augmented_resized_V2/train_cut2"

# Define the input size for the pre-trained models
input_size = (224, 224)

# Define batch size
batch_size = 32

# Create data generator with augmentation and preprocessing
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

# Load and preprocess training data
train_generator = datagen.flow_from_directory(
    data_dir,
    target_size=input_size,
    batch_size=batch_size,
    class_mode='sparse',
    subset='training'
)

# Load and preprocess validation data
validation_generator = datagen.flow_from_directory(
    data_dir,
    target_size=input_size,
    batch_size=batch_size,
    class_mode='sparse',
    subset='validation'
)

# Initialize pre-trained models
inception_resnet = InceptionResNetV2(weights='imagenet', include_top=False, input_shape=input_size + (3,))
xception = Xception(weights='imagenet', include_top=False, input_shape=input_size + (3,))
mobilenet = MobileNet(weights='imagenet', include_top=False, input_shape=input_size + (3,))
densenet = DenseNet121(weights='imagenet', include_top=False, input_shape=input_size + (3,))

# Feature extraction
def extract_features(model, generator):
    features = model.predict(generator)
    features = GlobalAveragePooling2D()(features)  # Apply global average pooling
    return features

# Extract features for training data
X_train_features = np.concatenate([
    extract_features(inception_resnet, train_generator),
    extract_features(xception, train_generator),
    extract_features(mobilenet, train_generator),
    extract_features(densenet, train_generator)
], axis=-1)

# Extract features for validation data
X_val_features = np.concatenate([
    extract_features(inception_resnet, validation_generator),
    extract_features(xception, validation_generator),
    extract_features(mobilenet, validation_generator),
    extract_features(densenet, validation_generator)
], axis=-1)

# Define number of classes
num_classes = len(train_generator.class_indices)

# Define your classifier on top of the extracted features
model = tf.keras.Sequential([
    Dense(128, activation='relu'),
    Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer=Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(X_train_features, train_generator.labels, epochs=5, validation_data=(X_val_features, validation_generator.labels))

# Evaluate the model on validation data
val_loss, val_accuracy = model.evaluate(X_val_features, validation_generator.labels)
print("Validation Accuracy:", val_accuracy)

# Predict on validation data
predictions = model.predict(X_val_features)
y_pred = np.argmax(predictions, axis=1)

# Print classification report
print(classification_report(validation_generator.labels, y_pred))
```
